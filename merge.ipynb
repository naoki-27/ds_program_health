# 取得データ

## 取得データ　睡眠

```python
import pandas as pd
```

```python
df_sleep = pd.read_csv('/Users/naoki/Lecture/ds_program_health/sleep_data.csv')
```

```python
df_sleep = df_sleep[::-1]
```

```python
df_sleep
```

```python
df_sleep['日付'] = pd.to_datetime(df_sleep['日付'])
```

```python
df_sleep
```

## 取得データ　health

```python
df_health = pd.read_csv('/Users/naoki/Lecture/ds_program_health/apple_health_export/health.csv')
```

```python
df_health
```

```python
df_health['datetime'] = pd.to_datetime(df_health['datetime'])
```

## 結合

```python
# 日付と datetime を共通のキーとして結合
df_local = pd.merge(df_sleep, df_health, left_on='日付', right_on='datetime', how='inner')
```

```python
# 必要なくなった datetime 列を削除
df_local = df_local.drop('datetime', axis=1)
```

```python
df_local
```

```python
df_local.to_csv('local.csv', index=False)
```

# スクレイピング

## スクレイピング　睡眠指数

```python
df_point = pd.read_csv('/Users/naoki/Lecture/ds_program_health/sleep_point.csv')
```

```python
df_point
```

```python
df_point['date'] = pd.to_datetime(df_point['date'])
```

## スクレイピング　気象庁

```python
df_weather = pd.read_csv('/Users/naoki/Lecture/ds_program_health/weather.csv')
```

```python
df_weather
```

```python
df_weather['day'] = pd.to_datetime(df_weather['day'])
```

```python
# 'day'と'date'を共通の列として結合
df_scraping = pd.merge(df_weather, df_point, left_on='day', right_on='date', how='inner')

# 共通の列を削除
df_scraping = df_scraping.drop(['date'], axis=1)
```

```python
df_scraping
```

```python
df_scraping.to_csv('scraping.csv', index=False)
```
