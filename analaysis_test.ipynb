```python
import sqlite3
import pandas as pd
```

```python
# SQLiteデータベースへの接続
conn = sqlite3.connect('sleep.sqlite')  # データベースのファイル名を指定してください

# SQLクエリを実行し、結果をデータフレームに変換
query = "SELECT * FROM sleep"  # テーブル名やクエリを適切に指定してください
df_sleep = pd.read_sql_query(query, conn)

# 接続を閉じる
conn.close()
```
